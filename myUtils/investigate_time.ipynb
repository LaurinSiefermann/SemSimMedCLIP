```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reading the CSV file
file_path = '/home/lsiefermann/open_clip_based_thesis/timings.csv'
timings_df = pd.read_csv(file_path)

# Calculating summary statistics
summary_statistics = timings_df.describe()
print(summary_statistics)

# Setting plot style
sns.set_style("whitegrid")

# Creating a box plot for each timing column
plt.figure(figsize=(15, 10))
sns.boxplot(data=timings_df)
plt.title('Distribution of Times for Different Processing Steps')
plt.ylabel('Time (seconds)')
plt.show()
```
